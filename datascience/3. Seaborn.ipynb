```python
df = df.drop(columns=['id_cliente'])
df.head()
```

```python
#separating the data
numerical_columns = df.select_dtypes(include='number').columns

categorical_columns = df.select_dtypes(include='object').columns

target_column = 'churn'
```

```python
#Remove the target columns
categorical_columns = categorical_columns.drop('churn')
```

## Numerical Variables

```python
df[numerical_columns].hist(layout=(1,3), figsize=(15,5));
```

```python
sns.set_theme(style='dark', context='notebook', palette='bright')

sns.histplot(x='meses_na_empresa', data=df, hue='contrato', multiple='stack')
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

for i, column in enumerate(numerical_columns):
    sns.histplot( x=column, data=df, hue='contrato', multiple='stack', ax=axs[i])


plt.show()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

for i, column in enumerate(numerical_columns):
    sns.histplot( x=column, data=df, hue='contrato', multiple='stack', ax=axs[i], bins =5)


plt.show()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5), tight_layout=True)

for i, column in enumerate(numerical_columns):
    sns.kdeplot(x=column, data=df, hue='churn', ax=axs[i], fill=True)
    legend = axs[i].get_legend()
    legend.remove()

labels = [text.get_text() for text in legend.get_texts()]

fig.legend(handles= legend.legend_handles, labels=labels, loc='upper center', ncols=2, title = 'Churn', bbox_to_anchor=(0.5,0.9))
fig.suptitle('Distribuição de variáveis numéricas por Churn\n\n', fontsize=20)
plt.show()
```

```python
#import formatter
from matplotlib.ticker import PercentFormatter

fig, axs = plt.subplots(nrows=4, ncols=4, figsize=(14,16),  sharey=True)

#Iterate through the columns and the axs
for i, column in enumerate(categorical_columns):
    h = sns.histplot(x=column, data=df, hue='churn', multiple='fill', ax=axs.flat[i], stat='percent', shrink=0.8)
    h.tick_params(axis='x', rotation=0)
    legend = h.get_legend()
    legend.remove()
    h.yaxis.set_major_formatter(PercentFormatter(1))
    h.set_ylabel('')

    #Iterate thorugh bar to apply the legend

    for bars in h.containers:
        h.bar_label(bars, label_type='center', labels=[f'{bar.get_height():.1%}' for bar in bars ])


#store labels
labels = [text.get_text() for text in legend.get_texts()]

#Change title and legend at figure level
fig.legend(handles= legend.legend_handles, labels=labels, loc='upper center', ncols=2, title = 'Churn', bbox_to_anchor=(0.5,0.95))
fig.suptitle('Distribuição de variáveis categóricas por Churn', fontsize=20)

fig.align_labels()

plt.subplots_adjust(wspace=0.3, hspace=0.3, top=0.9)

plt.show()
```
