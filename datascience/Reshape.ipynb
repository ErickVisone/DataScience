```python
import pandas as pd
```

```python
BASE = '../Data/reshaping.xlsx'

df = pd.read_excel(BASE)
df
```

```python
#Pivoting rows into columns
de_pivot = df.pivot(
    index = 'Estudante',
    columns= ['Categoria','Semana','Tópico'],
    values='Nota'
)

de_pivot
```

```python
#Chekc the inderx
de_pivot.index
```

```python
#Check the columns, Now I have a MultiIndex (Hierarchy)
de_pivot.columns
```

```python
de_pivot.stack(level = 0)
```

```python
de_pivot.stack(level=0).index
```

```python
de_pivot.stack(level=0).unstack(level=0)
```

```python
#Pivot Table --> pode ser feita agregacao enquanto a pivot nao pode.
df_pivot = df.pivot_table(
    index=["Estudante"],
    columns=["Tópico","Semana"],
    values="Nota"
)
```

```python
#This is doing the avg of both grades (Semana e Topico)
df_pivot
```

```python
#Pivot Table --> pode ser feita agregacao enquanto a pivot nao pode.
df_pivot2 = df.pivot_table(
    index=["Estudante"],
    columns=["Tópico","Semana"],
    values="Nota",
    aggfunc="sum" #count, mean, etc.. any aggragation works
)
```

```python
df_pivot2
```
