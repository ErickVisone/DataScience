## Segregation in Different Columns Types

```python
df.columns
```

```python
non_ordered_categorical_Columns = [
    'BusinessTravel',
    'Department',
    'EducationField',
    'Gender',
    'JobRole',
    'MaritalStatus',
    'OverTime'
]

ordered_categorical_columns = [
    'Education',
    'EnvironmentSatisfaction',
    'JobInvolvement',
    'JobLevel',
    'JobSatisfaction',
    'PerformanceRating',
    'RelationshipSatisfaction',
    'StockOptionLevel',
    'WorkLifeBalance'
]

target_columns = [
    'Attrition'
]

numerical_columns = [
    [column for column in df.columns if column not in (non_ordered_categorical_Columns + ordered_categorical_columns + target_columns)]
]
```

```python
numerical_columns
```

```python
for column in (non_ordered_categorical_Columns + ordered_categorical_columns):
    print(f"{column} : {df[column].unique()}")
```

```python
for column in non_ordered_categorical_Columns:
    df[column] = df[column].astype('category')
```

```python
df.info()
```

```python
for column in df.select_dtypes(include ='category'):
    print(f"{column}")
    print(f"{column}.cat.categories: {df[column].cat.categories}") 
    print(f"{column}.cat.ordered: {df[column].cat.ordered}")
    print(f"{column}.cat.codes: {df[column].cat.codes.unique()}")
    print()
```
